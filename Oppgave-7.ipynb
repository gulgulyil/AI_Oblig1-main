```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Use the inline backend for Matplotlib
%matplotlib inline
```

```python
# Read the CVS file
Grc_df = pd.read_csv('Grocery_dataset.csv')

# Drop columns with index 0 and 6 to create a new DataFrame
Grc_new_df = Grc_df.drop(Grc_df.columns[[0, 6]], axis=1)

# Create DataFrames for Supermarket Type1 and Type2
SupType_1 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type1']
SupType_2 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type2']

# Drop null values from both Item_MRP and Outlet_Size in SupType_1
SupType_1 = SupType_1.dropna(subset=['Item_MRP', 'Outlet_Size']) 
SupType_2 = SupType_2.dropna(subset=['Item_MRP'])

# Set the plot style to ggplot
plt.style.use('ggplot')

# Create a (2,1) subplot
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Box plot for SupType_1
sns.boxplot(data=SupType_1, x='Item_MRP', ax=axes[0])
axes[0].set_title('Item_MRP Box Plot for SupType_1')
axes[0].set_xlabel('Item MRP')

# Box plot for SupType_2
sns.boxplot(data=SupType_2, x='Item_MRP', ax=axes[1])
axes[1].set_title('Item_MRP Box Plot for SupType_2')
axes[1].set_xlabel('Item MRP')

# Adjust layout to prevet overlap
plt.tight_layout()

# Show the plots
plt.show()
```
