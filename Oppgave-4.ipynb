```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python

Grc_df = pd.read_csv('Grocery_dataset.csv')

Grc_df.head()
```

```python

Grc_df.info()

Grc_df.describe()
```

```python

unique_fat_content = Grc_df['Item_Fat_Content'].unique()
print("Unique Item Fat Content types:", unique_fat_content)
```

```python

num_unique_fat_content = Grc_df['Item_Fat_Content'].nunique()
print("Number of unique Item Fat Content types:", num_unique_fat_content)
```

```python

Grc_df['Item_Fat_Content'] = Grc_df['Item_Fat_Content'].str.lower()


unique_fat_content_fixed = Grc_df['Item_Fat_Content'].unique()
print("Unique Item Fat Content types after standardization:", unique_fat_content_fixed)
```

```python

Grc_df.loc[:, 'Item_Fat_Content'] = Grc_df['Item_Fat_Content'].replace({
    'low fat': 'low',
    'lf': 'low',
    'lowfat': 'low',
    'reg': 'regular',
    'regular': 'regular'
})



unique_fat_content_final = Grc_df['Item_Fat_Content'].unique()
print("Unique Item Fat Content types after cleaning:", unique_fat_content_final)
#Initially there were five values. But there were repeated values ​​among them. 
#After equating the repeated values, it turns out that there are actually two different values.
```
