```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

Grc_df = pd.read_csv('Grocery_dataset.csv')

Grc_new_df = Grc_df.drop(Grc_df.columns[[0, 6]], axis=1)

SupType_1 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type1']
SupType_2 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type2']

Grc_Concat_df = pd.concat([SupType_1, SupType_2])

# Create a multi-index for Grc_Concat_df using 'Outlet_Size' and 'Outlet_Location_Type'
Grc_Concat_df_multiindex = Grc_Concat_df.set_index(['Outlet_Size', 'Outlet_Location_Type'])

# Display the DataFrame with the multi-index
Grc_Concat_df_multiindex.head(10)
```
